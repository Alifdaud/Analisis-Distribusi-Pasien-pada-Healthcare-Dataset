```python
import pandas as pd

# Membaca dataset
df = pd.read_csv("/content/drive/MyDrive/Data_Mining/tugas_kuliah/healthcare_dataset.csv")

# Menampilkan 5 baris pertama
df.head()
```

```python
# 5 baris pertama
display(df.head())

# Info kolom & tipe data
print("Informasi Dataset:")
df.info()

# Statistik deskriptif kolom numerik
print("\nStatistik Deskriptif:")
display(df.describe())
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

print("Kolom Numerik:", list(numerical_columns))
print("Kolom Kategorikal:", list(categorical_columns))
```

```python
for col in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribusi Nilai: {col}')
    plt.xlabel(col)
    plt.ylabel('Frekuensi')
    plt.show()
```

```python
for col in categorical_columns:
    plt.figure(figsize=(10, 6))
    df[col].value_counts().head(3).plot.bar()
    plt.title(f'Top 3 kategori pada {col}')
    plt.xlabel(col)
    plt.ylabel('Frekuensi')
    plt.show()
```

```python
corr = df[numerical_columns].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Korelasi Antar Kolom Numerik')
plt.show()
```

1. Distribusi Blood Type

Tiga golongan darah terbanyak adalah A, A+, dan AB+, dengan frekuensi hampir sama (~7000).

Tidak ada dominasi signifikan antar tipe darah, distribusi relatif merata.

2. Distribusi Gender

Jumlah pasien laki-laki dan perempuan hampir seimbang, masing-masing sekitar 27.000+.

Ini menunjukkan dataset tidak bias ke salah satu gender.

3. Distribusi Medical Condition

Tiga kondisi medis terbanyak adalah Arthritis, Diabetes, dan Hypertension, masing-masing sekitar 9000+ pasien.

Hal ini wajar karena ketiga penyakit tersebut umum terjadi pada populasi usia menengah ke atas.

4. Distribusi Test Results

Hasil tes didominasi oleh kategori Abnormal, Normal, dan Inconclusive, dengan jumlah hampir sama (~18.000+).

Ini berarti proporsi pasien dengan hasil normal tidak jauh berbeda dengan yang mengalami abnormalitas.

5. Korelasi Antar Kolom Numerik

Korelasi antar variabel numerik (Age, Billing Amount, Room Number) sangat rendah (mendekati 0).

Artinya, usia pasien tidak berhubungan signifikan dengan jumlah tagihan maupun nomor kamar, dan antarvariabel numerik ini cenderung independen.
